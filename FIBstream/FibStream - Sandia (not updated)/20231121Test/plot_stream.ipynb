```python
rep = 2
scale = 37.8/2**12
# Generate data for a helix
x = path['x']*scale
y = path['y']*scale
z = path['z']/1.1e5*rep

# Create a 3D plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect([np.ptp(coord) for coord in [x, y, z]])
# Plot the helix
ax.plot(x, y, -1*z, label='FIB Trajectory', linewidth=0.1)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Add a legend
ax.legend()
ax.view_init(azim=70, elev=0)

# Show the plot
plt.show()
print('max is:', np.max(z))
```
