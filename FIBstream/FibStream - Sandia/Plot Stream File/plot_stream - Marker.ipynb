# Plot the path as a single picture

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
path = pd.read_csv('markerTest.csv')
```

```python
scale = 100/2**16
millingRate = 1.1e5 # micron / micro s

# Generate data for a helix
rep = path['rep'][0]
x = path['x']*scale
y = path['y']*scale
z = path['z']/millingRate*rep

plt.figure(figsize=(9, 9))
plt.scatter(x,y)
plt.xlabel('X Axis (micron)')
plt.ylabel('Y Axis (micron)')
plt.title('single L-Shape marker')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()

print("max z is: "+str(np.max(z))+" micron")
print("\nmin z is: "+str(np.min(z))+" micron")
```
