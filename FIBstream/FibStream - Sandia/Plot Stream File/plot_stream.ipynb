# Plot the path as a single picture

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
path = pd.read_csv('sil.csv')
```

```python
scale = 64/2**16
millingRate = 1.1e5 # micron / micro s

# Generate data for a helix
rep = path['rep'][0]
x = path['x']*scale
y = path['y']*scale
z = path['z']/millingRate*rep

# Create a 3D plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect([np.ptp(coord) for coord in [x, y, z]])
# Plot the helix
ax.plot(x, y, -1*z, label='FIB Trajectory', linewidth=0.1)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Add a legend
ax.legend()
ax.view_init(azim=70, elev=0)

# Show the plot
plt.show()
print('max is:', np.max(z))
```

# split the path as multiple pictures

```python
path1 = pd.read_csv('sil_00.csv')
path2 = pd.read_csv('sil_01.csv')
path3 = pd.read_csv('sil_02.csv')
```

```python
scale = 37.8/2**12
# Generate data for a helix
x1 = path1['x']*scale
y1 = path1['y']*scale
z1 = path1['z']/1.1e5*2

x2 = path2['x']*scale
y2 = path2['y']*scale
z2 = path2['z']/1.1e5*2

x3 = path3['x']*scale
y3 = path3['y']*scale
z3 = path3['z']/1.1e5*2

# Create a 3D plot
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect([np.ptp(coord) for coord in [x, y, z]])
# Plot the helix
ax.plot(x1, y1, -1*z1, label='FIB Trajectory', linewidth=0.1)
ax.plot(x2, y2, -1*(z2+3.6/3),  linewidth=0.1)
ax.plot(x3, y3, -1*(z3+3.6*2/3), linewidth=0.1)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Add a legend
ax.legend()
ax.view_init(azim=70, elev=0)

# Show the plot
plt.show()
```
